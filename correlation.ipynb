```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('dpe2.csv')
```

```python
# Convertir les colonnes de dates en timestamps
date_columns = ['Date_réception_DPE', 'Date_établissement_DPE', 'Date_visite_diagnostiqueur']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce').astype('int64') // 10**9
```

```python
# Calculer la matrice de corrélation
correlation_matrix = df.corr(numeric_only = True)

# Afficher la matrice de corrélation en utilisant seaborn
plt.figure(figsize=(30, 24))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```

```python
df3 = pd.read_csv('fusion.csv')
```

```python
# On va essayer de clean un peu, il y a trop de outliers

df3 = df3[df3["annee_construction"] > 1800]
df3 = df3[df3["annee_construction"] < 2030]
```

```python
# Convertir les colonnes de dates en timestamps
date_columns = ['date_etablissement_dpe']

for col in date_columns:
    df3[col] = pd.to_datetime(df3[col], errors='coerce').astype('int64') // 10**9
```

```python
# Calculer la matrice de corrélation
correlation_matrix3 = df3[df3["secteur_activite"] == 'Centre commercial'].corr(numeric_only = True)

# Afficher la matrice de corrélation en utilisant seaborn
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix3, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```
